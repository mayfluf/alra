```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Importa os dados reais das lojas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
```

```python
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
# Renomear as colunas para facilitar a manipulação
df_map = {
    'Produto': 'produto',
    'Categoria do Produto': 'categoria',
    'Preço': 'valor_venda',
    'Frete': 'custo_frete',
    'Avaliação da compra': 'avaliacao',
    'Tipo de pagamento': 'tipo_pagamento'
}

loja1 = loja1.rename(columns=df_map)
loja2 = loja2.rename(columns=df_map)
loja3 = loja3.rename(columns=df_map)
loja4 = loja4.rename(columns=df_map)
```

```python
lojas = {
    "Loja 1": loja1,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4
}
```

```python
# Verificação e limpeza básica dos dados para cada loja
for nome, df in lojas.items():
    print(f"\nAnalisando {nome}")
    print("Duplicatas:", df.duplicated().sum())
    print("Valores Nulos:\n", df.isnull().sum())

    # Define um índice mais simples baseado na coluna 'produto' se existir
    if 'produto' in df.columns:
        df.set_index('produto', inplace=True, drop=False)
```

```python
# Análise de métricas
faturamento_total = {}
avaliacao_media = {}
frete_medio = {}
produto_mais_vendido = {}
produto_menos_vendido = {}
categoria_popular = {}

for nome, df in lojas.items():
    faturamento_total[nome] = df['valor_venda'].sum()
    avaliacao_media[nome] = df['avaliacao'].mean()
    frete_medio[nome] = df['custo_frete'].mean()
    produto_mais_vendido[nome] = df['produto'].value_counts().idxmax()
    produto_menos_vendido[nome] = df['produto'].value_counts().idxmin()
    categoria_popular[nome] = df['categoria'].value_counts().idxmax()
```

```python
# Exibir métricas
print("\nFaturamento total:", faturamento_total)
print("\nAvaliação média:", avaliacao_media)
print("\nCusto médio do frete:", frete_medio)
print("\nProduto mais vendido:", produto_mais_vendido)
print("\nProduto menos vendido:", produto_menos_vendido)
print("\nCategoria mais popular:", categoria_popular)
```

```python

# 4. Distribuição Geográfica das Vendas por Loja
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()
for i, (nome, df) in enumerate(lojas.items()):
    scatter = axes[i].scatter(df['lon'], df['lat'], c=df['valor_venda'], cmap='plasma', alpha=0.6)
    axes[i].set_title(f"{nome} - Distribuição Geográfica")
    axes[i].set_xlabel("Longitude")
    axes[i].set_ylabel("Latitude")
fig.colorbar(scatter, ax=axes, orientation='horizontal', fraction=0.03, pad=0.07, label='Valor da Venda (R$)')
plt.suptitle("Comparativo Geográfico das Vendas por Loja", fontsize=16, y=0.98)
fig.subplots_adjust(hspace=0.3, wspace=0.2, top=0.90)
plt.savefig("comparativo_geografico_lojas.png")
```
